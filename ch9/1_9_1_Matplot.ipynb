```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = np.arange(10)
plt.plot(data)
```

```python
# subplots to figures  
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
# plt.plot uses last figure, 'k--' black dashed, fig reset after each cell
plt.plot(np.random.randn(50),'k--')
ax1.hist(np.random.randn(20), bins=20,color='k',alpha=.5)
ax2.scatter(np.arange(30), np.arange(30)*np.random.randn(30))
```

```python
# or other way, .subplots(array of subplots)
fig, axes = plt.subplots(2,3,sharex=True)
axes[0,0].plot(data)
plt.show()
```

```python
# subplots_adjust(left=None, bottom=None, right=None, top=None,
#  wspace=None, hspace=None)
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
     axes[i, j].hist(np.random.randn(500), bins=40, color='k', alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0)
```

```python
# style shorthand "<color><marker><linestyle>" like "ko--" or color,linestyle,marker indi
plt.plot(np.random.randn(40).cumsum(), 'ko--')
```

```python
plt.plot(np.random.randn(40).cumsum(), color="blue", marker="o" , linestyle="--") # or dashed
```

```python
# drawstyle='steps-post'
data = np.random.randn(30).cumsum()
plt.plot(data, 'k--', label='Default')
plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post')
plt.legend(loc='best')
```

### Ticks Labels Legends

```python
# pltsubplot -> xlim, xticks, xticklabels sets or gets
# AxesSubplot -> get_xlim, set_xlim

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())
ticks = ax.set_xticks([0, 250, 500,750,1000])
labels = ax.set_xticklabels(["one", "two", "three","four","five"],rotation=30, fontsize="small")
ax.set_title("My 1st plot")
ax.set_xlabel("Stages")
```

```python
# Set many props with ax.set(**props)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())
labels = ax.set_xticklabels(["one", "two", "three","four","five"],rotation=30, fontsize="small")

props = {
    "title" : "My 1st plot",
    "xlabel" : "Stages 2",
    "xticks": [0, 250, 500,750,1000]
}
ax.set(**props)
```

```python
# multiple lines in a plot then legend, figsize=lol
# pass label args to plot and then .legend
fig = plt.figure()

ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum(), 'k', label='one')
ax.plot(np.random.randn(1000).cumsum(), 'k--', label='two')
ax.plot(np.random.randn(1000).cumsum(), 'k.', label='three')
ax.legend(loc="best")
```

### Text arrow annotate

```python
# text, annotate, arrow
fig = plt.figure()

ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum(), 'k', label='one')
ax.plot(np.random.randn(1000).cumsum(), 'k--', label='two')
ax.plot(np.random.randn(1000).cumsum(), 'k.', label='three')
ax.legend(loc="best")

ax.text(20,20,"hiii")
ax.annotate("Here" , xy=(400,30) , arrowprops=dict(facecolor="black", headwidth=4,width=2))

# ax.annotate(label, xy=(arrowcords),
#  xytext=(text cords),
#  arrowprops=dict(facecolor='black', headwidth=4, width=2,
#  headlength=4),
#  horizontalalignment='left', verticalalignment='top'
```

### drawing shapes

```python
# Create shape then
# add shape with ax.add_patch(shape) only on ax

# plt.add_patch(circle) # not work
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

# cords in (0,1)
# (cords, r)
# (cords, l,b)
circle = plt.Circle((.4,.4), .2,color='g' , alpha = .4)
rectangle = plt.Rectangle((.8,.4),.5, .2,color='r' , alpha = .4)

# polygon cords
poly = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],color='k' , alpha = .4)

ax.add_patch(circle)
ax.add_patch(rectangle)
ax.add_patch(poly)


# save figure with, bbox_inches whitespace and dpi
plt.savefig("my_figure.svg", dpi=400, bbox_inches="tight")
```
