```python
import numpy as np 
import pandas as pd
```

```python
tips = pd.read_csv("../examples/tips.csv")
tips['tip_pct'] = tips['tip'] / tips['total_bill']
tips
```

```python
# top 5vals on col in df
def top(df, n=5,column='tip_pct'):
    return df.sort_values(by=column)[-n:]
top(tips,6)
```

```python
# group on smoker and apply top
tips.groupby('smoker').apply(top)
# heirarchial outer:grp name, inner:og inde
```

```python
# args to func after func name
tips.groupby(['smoker' , 'day']).apply(top,n=1,column='total_bill')
```

```python
# Supressing group keys heirarchial index
tips.groupby("smoker" , group_keys=False).apply(top)
# tips.groupby(["smoker", 'day'] , group_keys=False).apply(top)
```

### Bucket and Quantile 

```python
frame = pd.DataFrame({'data1': np.random.randn(1000),
     'data2': np.random.randn(1000)})
quartiles = pd.cut(frame.data1, 4)
quartiles
```

```python
# use it to group, get bucket wise stats
def get_stats(group):
    return { 'min'  : group.min() , 'max'  : group.max() ,'std'  : group.std(), 'mean'  : group.mean()}

grouping = pd.cut(frame.data2, 10,labels=False )
grouped = frame.data2.groupby(grouping)
grouped.apply(get_stats).unstack() # to take inner index to col
```
